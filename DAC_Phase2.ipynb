```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=3)
```

```python
x1
```

```python
x1_poly=poly.fit_transform(x1)
```

```python
x1_poly
```

```python
reg2=LinearRegression() #making polynmial model
reg2.fit(x1_poly,y1)
```

```python
plt.scatter(x1,y1,color='blue')
plt.plot(x1,linear_poly.predict(x1),color='red')
plt.xlabel("Row")
plt.ylabel("Day.Of.Week")
plt.title("Using Linear Regression")
```

```python
plt.scatter(x1,y1,color='blue')
plt.plot(x1,reg2.predict(poly.fit_transform(x1)),color='red')
plt.xlabel("Row")
plt.ylabel("Day.Of.Week")
plt.title("Using Polynomial Regression")
```
